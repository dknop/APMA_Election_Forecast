```python
corrMatrixURL = "https://raw.githubusercontent.com/TheEconomist/us-potus-model/master/data/state_correlation_matrix.csv"
corrMatrix = pd.read_csv(corrMatrixURL)
#corrMatrix.to_numpy()
#corrMatrix.iloc[[1,3],[1,3]]
#corrMatrix
BattlegroundStates = ['AZ','PA','FL','GA','MI','NV','TX','IA','NH','NC','OH','WI']
corrMatrixIndex = sorted([corrMatrix.columns.get_loc(colname) for colname in BattlegroundStates])
corrMatrixBattleStates = corrMatrix.iloc[corrMatrixIndex,corrMatrixIndex]


print(pres_2020_df['BidenAriz'][43])
vPA=0;vAZ=0;vFL = 0;vGA=0;vMI=0;vNV=0;vTX=0;vIA=0;vNH=0;vNC=0;vOH=0;vWI=0
varsList = [vAZ,vFL,vGA,vIA,vMI,vNV,vNH,vNC,vOH,vPA,vTX,vWI]
values = []
j = 1
for var in varsList:
  values.append(float(pres_2020_df_ordered.iloc[[43],[j]].to_numpy()))
  j=j+1

pAZ=values[0];pFL = values[1];pGA=values[2];pIA=values[3];pMI=values[4];pNV=values[5]
pNH=values[6];pNC=values[7];pOH=values[8];pPA=values[9];pTX=values[10];pWI=values[11]

probs = [pAZ,pFL,pGA,pIA,pMI,pNV,pNH,pNC,pOH,pPA,pTX,pWI]
probs = np.asarray(probs)
SDs = np.sqrt(probs*(1-probs)/(np.sqrt(BidenTradeVolume)))

vAZ=N(pAZ,SDs[0]);vFL=N(pFL,SDs[1]);vGA=N(pGA,SDs[2]);vIA=N(pIA,SDs[3]);vMI=N(pMI,SDs[4]);
vNV=N(pNV,SDs[5]);vNH=N(pNH,SDs[6]);vNC=N(pNC,SDs[7]);vOH=N(pOH,SDs[8]);vPA=N(pPA,SDs[9]);
vTX=N(pTX,SDs[10]);vWI=N(pWI,SDs[11])

corrMatrixBattleStates = corrMatrixBattleStates.to_numpy()
#corrplot0=plotcorr([vAZ,vFL,vGA,vIA,vMI,vNV,vNH,vNC,vOH,vPA,vTX,vWI])
correlate([vAZ,vFL,vGA,vIA,vMI,vNV,vNH,vNC,vOH,vPA,vTX,vWI],corrMatrixBattleStates)
#corrplot1=plotcorr([vAZ,vFL,vGA,vIA,vMI,vNV,vNH,vNC,vOH,vPA,vTX,vWI])

rvAZ=vAZ._mcpts; rvFL=vFL._mcpts; rvGA=vGA._mcpts; rvIA=vIA._mcpts; rvMI=vMI._mcpts;
rvNV=vNV._mcpts; rvNH=vNH._mcpts; rvNC=vNC._mcpts; rvOH=vOH._mcpts; rvPA=vPA._mcpts;
rvTX=vTX._mcpts; rvWI=vWI._mcpts

rand3 = np.vstack((rvAZ,rvFL,rvGA,rvIA,rvMI,rvNV,rvNH,rvNC,rvOH,rvPA,rvTX,rvWI))
df3 = pd.DataFrame(data=rand3).T
print(df3)

TrumpEVVotes = 126
BidenEVVotes = 223
State_EVVotes = [11,29,16,6,16,6,4,15,18,20,38,10]
BidenNumWins = 0
BidenElectoralVoteTotals = []
outcomes = [1,0,1,0,1,1,1,0,0,1,0,1]
BrierScoresArr = []
StatesPredictedCorrectArr = []
EVPredictedCorrectArr = []
CorrectStatesMatrix = np.zeros([12,10000])
for i in range(10000):
  x = random.uniform(0,1)
  BidenEVVotes = 223
  BidenNewVotes = 0
  BrierScore=0
  StatesPredictedCorrect = 0
  EVPredictedCorrect = 0

  for j in range(12):
    BrierScore = BrierScore + (rand3[j][i] - outcomes[j])**2
    x = random.uniform(0,1)
    if x < rand3[j][i]:
      BidenNewVotes = BidenNewVotes + State_EVVotes[j]
      if outcomes[j] == 1:
        StatesPredictedCorrect = StatesPredictedCorrect+1
        CorrectStatesMatrix[j][i]=1
        EVPredictedCorrect = EVPredictedCorrect + State_EVVotes[j]
    else:
      if outcomes[j] == 0:
        StatesPredictedCorrect = StatesPredictedCorrect+1
        CorrectStatesMatrix[j][i]=1
        EVPredictedCorrect = EVPredictedCorrect + State_EVVotes[j]
  BrierScore = BrierScore/12
  StatesPredictedCorrectArr.append(StatesPredictedCorrect)
  BrierScoresArr.append(BrierScore)
  BidenEVVotes = BidenEVVotes + BidenNewVotes
  BidenElectoralVoteTotals.append(BidenEVVotes)
  EVPredictedCorrectArr.append(EVPredictedCorrect)
  if BidenEVVotes > 270:
    BidenNumWins = BidenNumWins+1
CorrelMatrix = np.corrcoef(CorrectStatesMatrix)
print("Correlation Matrix:")
print(CorrelMatrix)
print("Mean EV Count: " + str(sum(BidenElectoralVoteTotals)/len(BidenElectoralVoteTotals)))
print("Standard Deviation of EV Count: " + str(np.std(BidenElectoralVoteTotals)))
print("Biden Percent Win Chance: "+ str(BidenNumWins/10000))
print("Average States Predicted Correct: " + str(sum(StatesPredictedCorrectArr)/len(StatesPredictedCorrectArr)))
print("Brier Score Average: " + str(sum(BrierScoresArr)/len(BrierScoresArr)))
print("Average Electoral Votes Predicted Correctly: " + str(sum(EVPredictedCorrectArr)/len(EVPredictedCorrectArr)))

BidenElectoralVoteTotals = np.asarray(BidenElectoralVoteTotals)
values,total_counts = np.unique(BidenElectoralVoteTotals,return_counts=True)
dems, counts_dems = np.unique(BidenElectoralVoteTotals[BidenElectoralVoteTotals>=270],return_counts=True)
reps, counts_reps = np.unique(BidenElectoralVoteTotals[BidenElectoralVoteTotals<=268],return_counts=True)
plt.figure(figsize = (10,7))
plt.bar(dems,counts_dems/np.sum(total_counts),label="Biden Win", color = 'blue')
plt.bar(reps,counts_reps/np.sum(total_counts),label="Trump Win", color = 'red')
plt.axvline(269, color = 'gray')
plt.title('Simulated Election Results from Betting Market Prices', fontsize = 18)
plt.xlabel("Electoral Votes Earned by Biden", fontsize = 14)
plt.ylabel ("Likelihood", fontsize = 14)
plt.legend()
plt.show()
```
